```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0,shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
knn_pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knncm=confusion_matrix(y_test,knn_pred)
sns.heatmap(knncm,annot=True)

knnac=accuracy_score(y_test,knn_pred)
print(knnac)

knncr=classification_report(y_test,knn_pred)
print(knncr)
```

```python
from sklearn.linear_model import LogisticRegression
logm=LogisticRegression()
logm.fit(x_train,y_train)
```

```python
logm_pred=logm.predict(x_test)

logmcm=confusion_matrix(y_test,logm_pred)
sns.heatmap(logmcm,annot=True)

logmac=accuracy_score(y_test,logm_pred)
print(logmac)

logmcr=classification_report(y_test,logm_pred)
print(logmcr)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt_pred=dt.predict(x_test)

dtcm=confusion_matrix(y_test,dt_pred)
sns.heatmap(dtcm,annot=True)

dtac=accuracy_score(y_test,dt_pred)
print(dtac)

dtcr=classification_report(y_test,dt_pred)
print(dtcr)
```
